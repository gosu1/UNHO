```python
import pandas as pd 
import matplotlib as mpl
from pandas import DataFrame 
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure 

matplotlib.rcParams['axes.unicode_minus'] = False

matplotlib.rcParams['font.family'] = "AppleGothic"

df = pd.read_csv('/Users/m/Desktop/graph/Corona.csv', encoding='cp949') 
fig = mpl.pyplot.gcf() 
plt.rcParams["figure.figsize"] = (14,4)
plot = df.plot() 
plot.set_xlabel("date")

plt.axvline(x=41 , color='red' , linestyle='--', linewidth=3) #구로구 콜센터 
plt.axvline(x=47 , color='red' , linestyle='--', linewidth=3) #세종정부청사
plt.axvline(x=21 , color='blue' , linestyle='--', linewidth=3) #31번 확진자(신천지)


plot.set_ylabel("확진자")
plt.title("1월20일부터 4월26일까지 확진자수") 
plt.xlabel("1.20 ~ 4.26 dates") 
plt.ylabel("확진자")

plt.show()
```
